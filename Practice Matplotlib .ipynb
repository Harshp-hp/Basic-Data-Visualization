```python
#importing the heart_disease Data file

heart_disease = pd.read_csv('heart-disease.csv')
heart_disease
```

```python
#showing first 10 rows 
heart_disease.head(10)
```

```python
#creating an histogram of the 'age' column with  bins

heart_disease.plot('age', kind='hist' ,bins = 50);
```

```python

# Call plot.hist() on the heart_disease DataFrame and toggle the
# "subplots" parameter to True

heart_disease.plot.hist(subplots=True);
```

```python
# Call the same line of code from above except change the "figsize" parameter
# to be (10, 30)

heart_disease.plot.hist(figsize=(10,30),subplots=True);
```

```python
heart_disease.plot.hist(figsize=(10,10));
```

```python
#creating a new DataFrame from the heart_disease Data

over_50 = heart_disease[heart_disease['age'] > 50];
over_50.head(10);

# create a plot 
#over_50.plot(kind='bar');
fig, ax = plt.subplots(figsize=(10,5));

#plot the Data
scatter = ax.scatter(x=over_50['age'],
                    y=over_50['chol'],
                    c = over_50['target']);

#customize the plot 
ax.set(title='Age and Cholestrol Levels',
      xlabel='Age',
      ylabel='Cholestrol');

#Adding a horizontal line for the cholestrol of mean()

ax.axhline(over_50['chol'].mean());

#Adding an legend 

ax.legend(*scatter.legend_elements(), title='Target');
```

```python
#create a plot 

fig, ax = plt.subplots(figsize=(10,6));

#plot the data 

scatter = ax.scatter(x= over_50['age'],
                    y= over_50['thalach'],
                    c= over_50['target']);

#customize the plot 

ax.set(title = 'Age and Thalach analysis',
         xlabel='Age',
         ylabel='Thalach');

#add the meanline()

ax.axhline(over_50['thalach'].mean(), 
          linestyle='--');

#adding a legend

ax.legend(*scatter.legend_elements(), title='Target');
```

```python
#adding 'cmap' in plotting to change the color of plot and adding color in axhline function.

#creating a plot
fig, ax = plt.subplots(figsize=(10,6));

#plot the data
scatter =  ax.scatter(x=over_50['age'],
                     y=over_50['thalach'],
                     c=over_50['target'],
                     cmap='winter');

#customize the plot 
ax.set(title = 'Analysis of Age and Thalach',
      xlabel='Age',
      ylabel='thalach');

#adding a meanline
ax.axhline(over_50['thalach'].mean(),
          linestyle='--',
          color='r');

#adding a legend

ax.legend(*scatter.legend_elements(), title = 'Target');
```

```python
#svaing the plot 
fig.savefig('Analysis of Age and Thalach.png');
```

```python
plt.style.use('seaborn')

#creating a subplots by using OO Method 

fig, (ax0,ax1) = plt.subplots(nrows=2,
                            ncols=1,
                        figsize=(10,10));

#plotting the first plot 'ax0'
#prepare a plot 
scatter = ax0.scatter(x=over_50['age'],
                     y=over_50['chol'],
                     c=over_50['target'],
                     cmap='winter');

#customize the plots
ax0.set(title='Analysis of Age and Cholestrol',
       xlabel='Age',
       ylabel='Cholestrol');

#Adding axhline 
ax0.axhline(over_50['chol'].mean(),
           linestyle='--',
           color='r');

#adding a legend 
ax0.legend(*scatter.legend_elements(), title='Target');


#plotting the 2nd plot 'ax2' 
#prepare data 
scatter = ax1.scatter(x=over_50['age'],
                     y=over_50['thalach'],
                     c=over_50['target'],
                     cmap='plasma');

#customize the plot 
ax1.set(title='Analysis of Age and Thalach',
       xlabel='Age',
       ylabel='Thalach');

#adding an legend 
ax1.legend(*scatter.legend_elements(), title='target');


#adding an line for mean
ax1.axhline(over_50['thalach'].mean(),
           linestyle='--',
           color='b');

#adding an title 

fig.suptitle('Heart Disease Analysis', fontsize='15', fontweight='bold');
```
